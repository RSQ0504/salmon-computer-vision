```python
def get_ids(dataset: dm.Dataset, subset: str):
    ids = []
    c = 0
    for item in dataset:
        if c > 8:
            break
        if item.subset == subset:
            ids += [item.id]
        c += 1

    return ids


#ids = get_ids(dataset, subset)
print("DatasetItem ids:", ids[:4])
```
